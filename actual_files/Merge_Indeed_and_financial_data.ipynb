```python
import pandas as pd
import pickle

indeed_ratings = pickle.load(open("s_p_indeed_ratings.p", "rb"))
sp_500_ebit_ebitda_prof = pickle.load(open("sp_500_ebit_ebitda_prof_div.p", "rb"))
```

```python

sp_500_ebit_ebitda_prof.head()
```

```python
indeed_ratings.head()
```

```python
indeed_ratings.rename(columns = {"Symbol":"ticker"}, inplace = True)
indeed_ratings.head()
```

```python
ratings_df_to_merge = pd.DataFrame()
ratings_df_to_merge["ticker"] = indeed_ratings["ticker"]
ratings_df_to_merge["company_ratings"] = indeed_ratings["company_ratings"]

ratings_and_financial_df = pd.merge(left = sp_500_ebit_ebitda_prof, 
                                    right = ratings_df_to_merge, on = "ticker",
                                   how = "left")
```

```python
ratings_and_financial_df.head()
```

```python
pickle.dump(ratings_and_financial_df, open("ratings_AND_fin_data_ebit_ebitda_div_profit.p", "wb"))
```
