```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import random
import time
import os

chromedriver = "/Applications/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(chromedriver)

# sample website- "https://www.indeed.com/cmp/Walmart/reviews"

def get_indeed_rating(company_name):
    
    web_address = "https://www.indeed.com/cmp/{}/reviews".format(company_name)
    print( "web_address is {}".format(web_address) )
    driver.get(web_address)
    time.sleep(2);
    ratings_head =driver.find_element_by_xpath("//*[@itemprop='ratingValue']")
    average_rating = ratings_head.text
    wait = random.uniform(8,11)
    time.sleep(wait)

    
    return(average_rating)

def scrape_list(company_list):
    company_ratings = {}
    errors = []
    for company in company_list:
        print("company{}".format(company))
        try: 
            rating = get_indeed_rating(company)
            company_ratings[company] = rating
        except:
           print("error")  # I could create a log of errors
           errors.append(company)
        results = (company_ratings, errors)
    return (results)
            
```

```python
import re
#make the company names into strings that are readable in the format that Indeed uses.



#if last has "Inc", "Plc" "Corp" remove that word and remove everything that comes after it
#remove trailing white_space.
# add dashes in place of white space



def reformat(c_name):
    c_name, sep, tail = c_name.partition("Inc")
    c_name, sep, tail = c_name.partition("Plc")
    c_name, sep, tail = c_name.partition("Corp")
    c_name = c_name.rstrip()
    c_name = re.sub(r"[^\w\s]", '', c_name)
    c_name = re.sub(r"\s+", '-', c_name)

    
    return(c_name)

print(reformat("hi mike  Corp        k"))
    
# handy "urlify function here:  https://stackoverflow.com/questions/1007481/how-do-i-replace-whitespaces-with-underscore-and-vice-versa
# last two regex functions above come frome that link
```
