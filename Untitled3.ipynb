```python
mask_prof = (
    (fundamentals_transposed["indicator"] == "EBIT" )| 
    (fundamentals_transposed["indicator"] == "EBITDA" )|
    (fundamentals_transposed["indicator"] == "Net Profit" )|
    (fundamentals_transposed["indicator"] == "Dividends" ) 
    )

prof_div_ebit_df = fundamentals_transposed[mask_prof]
#prof_div_ebit_df.sort_values(by = "indicator
#                             ")
prof_div_ebit_df
```

```python
sp_500_ebit_ebitda_prof_div = pd.merge(left = s_p_comp_list, right = prof_div_ebit_df, 
                                       how = "left", on = "ticker")


sp_500_ebit_ebitda_prof_div
```
